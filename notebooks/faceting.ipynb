````python
# Your code below (You need to make the appropriate imports of matplotlib and numpy)
import matplotlib.pyplot as plt
import numpy as np
jjdata="0.71  0.63  0.85  0.44 \
0.61  0.69  0.92  0.55 \
0.72  0.77  0.92  0.60 \
0.83  0.80  1.00  0.77 \
0.92  1.00  1.24  1.00 \
1.16  1.30  1.45  1.25 \
1.26  1.38  1.86  1.56 \
1.53  1.59  1.83  1.86 \
1.53  2.07  2.34  2.25 \
2.16  2.43  2.70  2.25 \
2.79  3.42  3.69  3.60 \
3.60  4.32  4.32  4.05 \
4.86  5.04  5.04  4.41 \
5.58  5.85  6.57  5.31 \
6.03  6.39  6.93  5.85 \
6.93  7.74  7.83  6.12 \
7.74  8.91  8.28  6.84  \
9.54 10.26  9.54  8.73 \
11.88 12.06 12.15  8.91 \
14.04 12.96 14.85  9.99 \
16.20 14.67 16.02 11.61"
y = jjdata.replace('  ',' ').split(' ')
y = list(map(float,y))
y = np.array(y).reshape(21,4)

quarter_one = [quart1[0] for quart1 in y]
quarter_two = [quart2[1] for quart2 in y]
quarter_three = [quart3[2] for quart3 in y]
quarter_four = [quart4[3] for quart4 in y]
xyears = list(range(1960,1981,1))
fig = plt.figure(figsize = (6, 10))

fig.suptitle("Johnson and Johnson Quarterly Earnings", fontsize=14)
          
# Creating first subplot
plt.subplot(4, 1, 1)
plt.title("First Quarter Earnings")
plt.xlabel("Year")
plt.ylabel("Earnings")

plt.xticks(xyears, xyears, rotation ='vertical')
plt.locator_params(axis="x", nbins=10.5)
plt.plot(xyears,quarter_one)

# Creating second  subplot
plt.subplot(4, 1, 2)
plt.title("Second Quarter Earnings")
plt.xlabel("Year")
plt.ylabel("Earnings")
plt.xticks(xyears, xyears, rotation ='vertical')
plt.locator_params(axis="x", nbins=10.5)
plt.plot(xyears,quarter_two)

# Creating third  subplot
plt.subplot(4, 1, 3)
plt.title("Third Quarter Earnings")
plt.xlabel("Year")
plt.ylabel("Earnings")
plt.xticks(xyears, xyears, rotation ='vertical')
plt.locator_params(axis="x", nbins=10.5)
plt.plot(xyears,quarter_three)

# Creating fourth  subplot
plt.subplot(4, 1, 4)
plt.title("Fourth Quarter Earnings")
plt.xlabel("Year")
plt.ylabel("Earnings")
plt.xticks(xyears, xyears, rotation ='vertical')
plt.locator_params(axis="x", nbins=10.5)
plt.plot(xyears,quarter_four)


plt.tight_layout()
fig.subplots_adjust(hspace=1)
plt.show()



````

For this challenge:
* Facet the plots on the number of cylenders each vechical has
* Create a single row with with the number of correct plots
* Use a scatter plot to show the relation between horse power and quarter mile time
* On the x axis plot the horse power
* On the y axis plot the the quarter mile time (qsec)
* Each plot should have a title saying the number of cylinders
* Give each axis an appropriate name

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

cars_data = pd.read_csv('../data/cars.csv', sep='|')
cars_data.rename(columns={'Unnamed: 0': 'make_model'}, inplace=True)
#cars_data.head()

fourcyl=cars_data[cars_data['cyl']==4]
sixcyl=cars_data[cars_data['cyl']==6]
eightcyl=cars_data[cars_data['cyl']==8]


fig, ax = plt.subplots(1, 3)



# Creating first subplot
fourhp = fourcyl['hp'].tolist()
fourqsec = fourcyl['qsec'].tolist()
plt.subplot(1, 4, 1)
plt.title("4 Cylinder Cars")
plt.xlabel("Horse Power")
plt.ylabel("QSec")
plt.scatter(fourhp,fourqsec)

# Creating 2nd subplot
sixhp = sixcyl['hp'].tolist()
sixqsec = sixcyl['qsec'].tolist()
plt.subplot(1, 4, 2)
plt.title("6 Cylinder Cars")
plt.xlabel("Horse Power")
plt.ylabel("QSec")
plt.scatter(sixhp,sixqsec)

# Creating 3rd subplot
eighthp = eightcyl['hp'].tolist()
eightqsec = eightcyl['qsec'].tolist()
plt.subplot(1, 4, 3)
plt.title("8 Cylinder Cars")
plt.xlabel("Horse Power")
plt.ylabel("QSec")
plt.scatter(eighthp,eightqsec)

plt.tight_layout()
fig.subplots_adjust(hspace=1)
````

````python
fig = plt.figure(figsize = (8, 12))
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle("\n".join(["Qsec Distribution by HP"]), y=0.98)

# Set common labels
fig.text(0.5, 0.04, 'HP', ha='center', va='top')
fig.text(0.06, 0.5, 'QSEC', ha='right', va='top', rotation='vertical')


#plt.subplot(1, 3, 1)
ax1.title.set_text("4 cylinder")
ax1.scatter(fourhp, fourqsec)

#plt.subplot(1, 3, 2)
ax2.title.set_text("6 cylinder")
ax2.scatter(sixhp, sixqsec)

#plt.subplot(1, 3, 3)
ax3.title.set_text("8 cylinder")
ax3.scatter(eighthp, eightqsec)

plt.tight_layout()
fig.subplots_adjust(hspace=1)

plt.savefig('../images/facet_cyl_cars.png')
````
